```python
df_train["Title"]=df_train["Name"].apply(lambda x:x.split(", ")[1].split(".")[0])
df_train["Title"].value_counts()
```

```python
L=["Mr","Mrs","Miss","Master","Dr","Rev"]
df_train["Name_Title"]=df_train["Title"].apply(lambda x:x if x in L else "other")
```

```python
(df_train["Name_Title"]=="other").sum()
```

```python
df_train.info()
```

### filling missing values

```python
df_train["Age"].fillna(df_train["Age"].mean(),inplace=True)
df_train["Cabin"]=df_train["Cabin"].isnull().apply(lambda x:0 if x==True else 1)
df_train["Embarked"].fillna(df_train["Embarked"].mode().iloc[0],inplace=True)
```

```python
df_train.info()
```

### drop unnecessary columns

```python
df_train_new=df_train.drop(columns=["PassengerId","Name","Ticket","Title"],axis=1)
df_train_new
```

### one hot encoding

```python
from sklearn.preprocessing import OneHotEncoder
OHE=OneHotEncoder()
OHE.fit(df_train_new[["Sex","Embarked","Name_Title"]])
A=OHE.transform(df_train_new[["Sex","Embarked","Name_Title"]]).toarray()
OHE.get_feature_names()
```

```python
B=pd.DataFrame(data=A,columns=OHE.get_feature_names())
# drop x0_male,x1_S,x2_other
B.drop(columns=["x0_male","x1_S","x2_other"],axis=1,inplace=True)
B
```

```python
df_test.info()
```

```python
df_test["title1"]=df_test["Name"].apply(lambda x:x.split(",")[1].split(".")[0])
L=df_test["title1"].value_counts().index[:6]
df_test["title1"]=df_test["title1"].apply(lambda x:x if x in L else "other")
```

# model building

```python
from sklearn.linear_model import LogisticRegression
LC = LogisticRegression()
LC.fit(X_train,y_train)
LC.score(X_train,y_train)
```

```python
from sklearn.tree import DecisionTreeClassifier
DT =DecisionTreeClassifier(min_samples_leaf=3,min_samples_split=3,criterion='entropy', max_depth=5)
DT.fit(X_train,y_train)
DT.score(X_train,y_train)
```

```python
yp=DT.predict(X_test)
df_test["Survived"]=yp
df_test[['PassengerId','Survived']].to_csv('submit2.csv',index=False)
```
